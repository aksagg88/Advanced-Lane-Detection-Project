```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
import cv2
import numpy as np


# Create a black image
img = np.zeros((720,1280,3),np.uint8)

# pts
pts = np.array([[80,710],[1200,710],[640,360]],np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(0,255,255))

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'(700,80)',(80,680), font, 1,(255,255,255),2,cv2.LINE_AA)
cv2.putText(img,'(700,1200)',(1100,680), font, 1,(255,255,255),2,cv2.LINE_AA)
cv2.putText(img,'(360,640)',(640,340), font, 1,(255,255,255),2,cv2.LINE_AA)

plt.imshow(img)
```

```python
lb = [80,710]
rb = [1200,710]
ap = [640,360]

## fit lines (y = Ax + B)
fitL = np.polyfit((lb[0], ap[0]), (lb[1], ap[1]), 1)
fitR = np.polyfit((rb[0], ap[0]), (rb[1], ap[1]), 1)
fitB = np.polyfit((lb[0], rb[0]), (lb[1],rb[1]), 1)

print(fitL, fitR, fitB)

plt.plot([lb[0],rb[0]],[lb[1],rb[1]],'r--',lw=10)
plt.imshow(img)
```
